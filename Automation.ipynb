```python
# Task 3
# Task Automation with Python Scripts

import os
import shutil
import request

source_folder = "source_images"
destination_folder = "moved_images"

os.makedirs(source_folder, exist_ok=True)
os.makedirs(destination_folder, exist_ok=True)


image_urls = [
    "https://picsum.photos/300/200.jpg",
    "https://picsum.photos/400/300.jpg",
    "https://picsum.photos/500/300.jpg"
]


for i, url in enumerate(image_urls):
    response = requests.get(url)
    filename = f"image_{i+1}.jpg"

    with open(os.path.join(source_folder, filename), "wb") as file:
        file.write(response.content)

    print(f"Downloaded: {filename}")


for filename in os.listdir(source_folder):
    if filename.endswith(".jpg"):
        shutil.move(
            os.path.join(source_folder, filename),
            os.path.join(destination_folder, filename)
        )
        print(f"Moved: {filename}")

print("✅ All images downloaded and moved successfully!")

```
